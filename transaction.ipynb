# 13.What percentage of customers are repeat buyers?


```python
tran = df.groupby("cust_id").size()
repeat_buyers = tran[tran > 1]
num_repeat = repeat_buyers.count()
total = df["cust_id"].nunique()
per = (num_repeat / total) * 100
print(per)
```
